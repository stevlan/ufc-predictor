```python
import requests
from bs4 import BeautifulSoup 
import string
import pandas as pd
import csv
```

```python
fightsDF = pd.read_csv("fightStats.csv")
fightersDF = pd.read_csv("fighterStats.csv")


# fightsDF["Knockdown Differential"] = fightsDF["Knockdowns (R)"]-fightsDF["Knockdowns (B)"]
fightsDF = fightsDF.drop(columns=["Knockdowns (R)", "Knockdowns (B)"])

# fightsDF["Significant Strikes Differential"] = fightsDF["Significant Strikes (R)"]-fightsDF["Significant Strikes (B)"]
fightsDF = fightsDF.drop(columns=["Significant Strikes (R)", "Significant Strikes (B)"])

# fightsDF["Significant Strikes Percentage Differential"] = fightsDF["Significant Strikes Percentage (R)"]-fightsDF["Significant Strikes Percentage (B)"]
fightsDF = fightsDF.drop(columns=["Significant Strikes Percentage (R)", "Significant Strikes Percentage (B)"])

# fightsDF["Total Strikes Differential"] = fightsDF["Total Strikes (R)"]-fightsDF["Total Strikes (B)"]
fightsDF = fightsDF.drop(columns=["Total Strikes (R)", "Total Strikes (B)"])

# fightsDF["Takedowns Differential"] = fightsDF["Takedowns (R)"]-fightsDF["Takedowns (B)"]
fightsDF = fightsDF.drop(columns=["Takedowns (R)", "Takedowns (B)"])

# fightsDF["Takedowns Percentage Differential"] = fightsDF["Takedowns Percentage (R)"]-fightsDF["Takedowns Percentage (B)"]
fightsDF = fightsDF.drop(columns=["Takedowns Percentage (R)", "Takedowns Percentage (B)"])

# fightsDF["Submission Attempts Differential"] = fightsDF["Submission Attempts (R)"]-fightsDF["Submission Attempts (B)"]
fightsDF = fightsDF.drop(columns=["Submission Attempts (R)", "Submission Attempts (B)"])

# fightsDF["Reversals Differential"] = fightsDF["Reversals (R)"]-fightsDF["Reversals (B)"]
fightsDF = fightsDF.drop(columns=["Reversals (R)", "Reversals (B)"])

# fightsDF["Control Time (seconds) Differential"] = fightsDF["Control Time (seconds) (R)"]-fightsDF["Control Time (seconds) (B)"]
fightsDF = fightsDF.drop(columns=["Control Time (seconds) (R)", "Control Time (seconds) (B)"])

reach = []
height = []
# weight = [] not good indicator. some fighters may be fighting in a diff weight class for one fight so normal weight isnt accurate predictor
ssPerMin = []
ssAcc = []
ssAbs = []
ssDef = []
avgTD = []
tdAcc = []
tdDef = []
avgSub = []
for index, row in fightsDF.iterrows():

    redReach = fightersDF.loc[fightersDF["Name"] == row["Red"], "Reach (inches)"].values[0]
    blueReach = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Reach (inches)"].values[0]
    reachDifference = redReach - blueReach
    reach.append(reachDifference)

    redHeight = fightersDF.loc[fightersDF["Name"] == row["Red"], "Height (inches)"].values[0]
    blueHeight = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Height (inches)"].values[0]
    heightDifference = redHeight - blueHeight
    height.append(heightDifference)

    # redWeight = fightersDF.loc[fightersDF["Name"] == row["Red"], "Weight (pounds)"].values[0]
    # blueWeight = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Weight (pounds)"].values[0]
    # weightDifference = redWeight - blueWeight
    # weight.append(weightDifference)

    redSSPerMin = fightersDF.loc[fightersDF["Name"] == row["Red"], "Significant Strikes Landed per Minute"].values[0]
    blueSSPerMin = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Significant Strikes Landed per Minute"].values[0]
    ssPerMinDifference = redSSPerMin - blueSSPerMin
    ssPerMin.append(ssPerMinDifference)

    redSSAcc = fightersDF.loc[fightersDF["Name"] == row["Red"], "Significant Striking Accuracy"].values[0]
    blueSSAcc = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Significant Striking Accuracy"].values[0]
    ssAccDifference = redSSAcc - blueSSAcc
    ssAcc.append(ssAccDifference)

    redSSAbs = fightersDF.loc[fightersDF["Name"] == row["Red"], "Significant Strikes Absorbed per Minute"].values[0]
    blueSSAbs = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Significant Strikes Absorbed per Minute"].values[0]
    ssAbsDifference = redSSAbs - blueSSAbs
    ssAbs.append(ssAbsDifference)

    redSSDef = fightersDF.loc[fightersDF["Name"] == row["Red"], "Significant Strike Defence (the percentage of opponents strikes that did not land)"].values[0]
    blueSSDef = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Significant Strike Defence (the percentage of opponents strikes that did not land)"].values[0]
    ssDefDifference = redSSDef - blueSSDef
    ssDef.append(ssDefDifference)

    redAvgTD = fightersDF.loc[fightersDF["Name"] == row["Red"], "Average Takedowns Landed per 15 minutes"].values[0]
    blueAvgTD = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Average Takedowns Landed per 15 minutes"].values[0]
    avgTDDifference = redAvgTD - blueAvgTD
    avgTD.append(avgTDDifference)

    redTDAcc = fightersDF.loc[fightersDF["Name"] == row["Red"], "Takedown Accuracy"].values[0]
    blueTDAcc = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Takedown Accuracy"].values[0]
    tdAccDifference = redTDAcc - blueTDAcc
    tdAcc.append(tdAccDifference)

    redTDDef = fightersDF.loc[fightersDF["Name"] == row["Red"], "Takedown Defense (the percentage of opponents TD attempts that did not land)"].values[0]
    blueTDDef = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Takedown Defense (the percentage of opponents TD attempts that did not land)"].values[0]
    tdDefDifference = redTDDef - blueTDDef
    tdDef.append(tdDefDifference)

    redAvgSub = fightersDF.loc[fightersDF["Name"] == row["Red"], "Average Submissions Attempted per 15 minutes"].values[0]
    blueAvgSub = fightersDF.loc[fightersDF["Name"] == row["Blue"], "Average Submissions Attempted per 15 minutes"].values[0]
    avgSubDifference = redAvgSub - blueAvgSub
    avgSub.append(avgSubDifference)



fightsDF["Reach (inches) Differential"] = reach
fightsDF["Height (inches) Differential"] = height
# fightsDF["Weight (pounds) Differential"] = weight
fightsDF["Significant Strikes Landed per Minute Differential"] = ssPerMin
fightsDF["Significant Striking Accuracy Differential"] = ssAcc
fightsDF["Significant Strikes Absorbed per Minute Differential"] = ssAbs
fightsDF["Significant Strike Defence (the percentage of opponents strikes that did not land) Differential"] = ssDef
fightsDF["Average Takedowns Landed per 15 minutes Differential"] = avgTD
fightsDF["Takedown Accuracy Differential"] = tdAcc
fightsDF["Takedown Defense (the percentage of opponents TD attempts that did not land) Differential"] = tdDef
fightsDF["Average Submissions Attempted per 15 minutes"] = avgSub

fightsDF = fightsDF.drop(columns=["Loser"])
fightsDF = fightsDF.drop(columns=["Red"])
fightsDF = fightsDF.drop(columns=["Blue"])

fightsDF.to_csv("differentials.csv")




```
